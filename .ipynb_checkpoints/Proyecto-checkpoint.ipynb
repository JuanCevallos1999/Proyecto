```python
import pandas as pd
from math import pi
from bokeh.io import output_file, show, output_notebook #Permite presentar una imagen formato html5
from bokeh.layouts import column #Permite almacenar en columnas
from bokeh.plotting import figure  #Permite crear un objeto vacio
from bokeh.palettes import inferno #Paleta de colores
from bokeh.palettes import *  
from bokeh.transform import cumsum #Permite sumar los angulos
output_notebook()
data = pd.read_csv("data/defunciones.csv", sep=",", encoding = "ISO 8859-1") #Lee el csv con el encoding que permite la codificacion del lenguaje latino
```

```python
data
#Presentamos  el csv leido por el pandas
```

```python
Sexo = ["Hombre","Indeterminado","Mujer"]
#Creamos un encabezado que sera utilizado posteriormente como rango
```

```python
contadorsexo = data.groupby(["sexo"])["sexo"].count() #Permite agrupar en una variable todo lo almacenado en la variable sexo
```

```python
output_file("Sexo.html") #Arhivo html que se creará la gráfica de barras con el atributo sexo
a = figure(x_range = Sexo, plot_height= 500, plot_width=700, title="Sexo", toolbar_location=None, tools="") #Creamos una 'ventana', con las especificaciónes necesarias
a.vbar(x=Sexo,top=contadorsexo,width=0.3,line_color ="black") #Creamos cada barra con los atributos que solicita
show(a) #presentamos el objeto
output_notebook()
```

```python
#Grafica pastel que presenta las defunciones por provincia del año 2014
color = ["E82424","E83624","E84B24","E85F24","E87A24","E88F24","E8AA24","E8CD24","D6E824","C4E824","86E824","71E824","68E824","4BE824"
        ,"24E833","24E859","24E87D","24E889","24E8B3","24E8E2","24CAE8","24B0E8","2489E8","2451E8"]
x = data.groupby(["prov_fall"])["prov_fall"].count() #Agrupamos los datos del atributo prov_fall
data = pd.Series(x).reset_index(name='value').rename(columns={'index':'prov_fall'}) 
data['angle'] = data['value']/data['value'].sum() * 2*pi #Calcula el angulo que le corresponde en el circulo
data['color'] = inferno(24) #Agrupa colores
p = figure(plot_height=350, title="Provincias Grafico Pastel", toolbar_location=None, #Crea la 'ventana'
           tools="hover", tooltips="@prov_fall: @value", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.3, #dimensiones del circulo
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'), #suma de angulos 
        line_color="black", fill_color='color', legend="prov_fall", source=data) #color de la linea, pintar, recorrido de la variable en data
#No presentamos el menu
p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p) #Mostramos el objeto
output_notebook()
```

```python
 contadorMes=data.groupby(["mes_fall"])["mes_fall"].count()  #Agrupamos los datos de la variable mes_insc que son el mes inscrito
```

```python
Meses=["Abril","Agosto","Diciembre","Enero","Feberero","Julio","Junio","Marzo","Mayo","Noviembre","Octubre","Septiembre"] #Encabezado
```

```python
output_file("Mes.html")
b = figure(x_range = Meses, plot_height= 500, plot_width=700, title="Mes", toolbar_location=None, tools="") #Creamos la 'ventana'
b.vbar(x=Meses,top=contadorMes,width=0.3) #Enviamos los parametros a cada barra
b.xgrid.grid_line_color = None
b.y_range.start=0
show(b)
output_notebook()
```

```python
contadorEtnia =data.groupby(["etnia"])["etnia"].count()
```

```python
factors = ["AfroAmericano", "Blanca", "Ignoro", "Indigenas", "Mestiza", "Montubia", "Mulata", "Negra","Otra"]
x= contadorEtnia
b = figure(y_range=factors)
b.circle(x, factors, size=15, fill_color="red", line_color="black", line_width=3)
b.line(x,factors,color="black")
show(b)
output_notebook()
```

```python
contadorInstruccion = data.groupby(["niv_inst"])["niv_inst"].count()
```

```python
output_file("BarrasOrdenadas.html")

Instruccion = ['Alfabetización', 'Post-Bachillerato', 'E.Básica', 'E.Media', 'Ninguno', 'PostGrado','Primaria','Se ignora','Secundaria','Superior']
sorted_instruccion = sorted(Instruccion, key=lambda x: contadorInstruccion[Instruccion.index(x)])
p = figure(x_range=sorted_instruccion, plot_height=350, title="Nivel de instrucción",
           toolbar_location=None, tools="")
p.vbar(x=Instruccion, top=contadorInstruccion, width=0.9)
p.xgrid.grid_line_color = None
p.y_range.start = 0
show(p)
output_notebook()
```
